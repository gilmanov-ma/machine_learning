```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.tsa.api as smt
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import pmdarima as pm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

# Модель SARIMAX

```python
df_clear = pd.read_excel('C:\\Users\\Marat Gilmanov\\CMPRO Dropbox\\Аналитика\\Доклады и статьи\\11. Доклады в Союзцементе\\2024\\01 (08 февраля 2024)\\Прогнозы\\Ряды для расчетов.xlsx', sheet_name="Потребление цемента" )
```

```python
df_clear['date'] = pd.to_datetime(dict(year=df_clear['Год'], month=df_clear['Месяц'], day=1))
```

```python
df_clear
```

## Разложение временного ряда на компоненты

```python
df = df_clear
```

```python
decompose_result = seasonal_decompose(df['Потребление'], model="multiplicative", period=12)

trend = decompose_result.trend
seasonal = decompose_result.seasonal
residual = decompose_result.resid
decompose_result.plot();
```

```python
smt.graphics.plot_acf(df['Потребление'], lags=12,  alpha=0.5)
smt.graphics.plot_pacf(df['Потребление'], lags=12, alpha=0.5);
```

## Проверка временного ряда на стационарность

```python
# Perform ADF test
result = adfuller(df['Потребление'])
print('ADF test:\tp-value: {:.3f}'.format(result[1]))
```

### Дифференцирование по сезонной составляющей

```python
df['Потребление Season'] = df['Потребление'] - df['Потребление'].shift(12)
df.dropna(subset = ["Потребление Season"], inplace=True)
df
```

```python
smt.graphics.plot_acf(df['Потребление Season'], lags=12,  alpha=0.5)
smt.graphics.plot_pacf(df['Потребление Season'], lags=12, alpha=0.5);
```

```python
result = adfuller(df['Потребление Season'])
print('ADF test:\tp-value: {:.3f}'.format(result[1]))
```

### Вычитаем первую разность

```python
df['Потребление Diff'] = df['Потребление Season'] - df['Потребление Season'].shift(1)
df.dropna(subset = ["Потребление Diff"], inplace=True)
df
```

```python
smt.graphics.plot_acf(df['Потребление Diff'], lags=12,  alpha=0.5)
smt.graphics.plot_pacf(df['Потребление Diff'], lags=12, alpha=0.5);
```

```python
result = adfuller(df['Потребление Diff'])
print('ADF test:\tp-value: {:.10f}'.format(result[1]))
```

```python
df['Потребление Diff'].plot(figsize=(5,3))
```

## Подбираем Параметры PDQ (pdq)

```python
arima_model = pm.auto_arima(df_clear['Потребление'],
                           start_p=0, start_q=0,
                           test='adf',
                           max_p=5, max_q=5, m=12,
                           start_P=0, start_Q=0,
                           max_P=2, max_Q=2,
                           seasonal=True,
                           d=1, D=1,
                           trace=True,
                           error_action='ignore',
                           suppress_warnings=True,
                           stepwise=False)
```

## Обучаем модель на выборке до 2023 года

```python
best_model_2023 = sm.tsa.statespace.SARIMAX(df_clear[df_clear['date'] < '2023-01-01']['Потребление'], order=(1, 1, 1), seasonal_order=(0, 1, 1, 12)).fit(disp=False)
print(best_model_2023.summary())
```

```python
best_model_2023.plot_diagnostics(figsize=(10,7));
```

## Получаем прогнозы на 2023 год

```python
forecast_values = best_model_2023.get_forecast(steps = 12)
conf_int = forecast_values.conf_int()
conf_int['date'] = pd.date_range(start='2022-12-01', periods=12, freq='m') + pd.DateOffset(1)
conf_int
```

```python
evaluation_results = pd.DataFrame({'r2_score': r2_score(df_clear[(df_clear['date'] >= '2023-01-01') & (df_clear['date'] < '2024-01-01')]['Потребление'], forecast_values.predicted_mean)}, index=[0])
evaluation_results
```

```python
future_dates = pd.date_range(start='2022-12-01', periods=12, freq='m') + pd.DateOffset(1)
forecast_df = pd.DataFrame({'date': future_dates, 'Потребление': forecast_values.predicted_mean})
forecast_df
```

```python
df_new = pd.concat([df_clear[df_clear['date'] < '2023-01-01'], forecast_df])
df_new.to_excel('df.xlsx')
```

```python
#Plot the data
ax = df_new.plot(y='Потребление', figsize = (10, 5), x='date',legend = True, color='purple')
plt.show();
```
